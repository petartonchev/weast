```python
import pandas as pd
import wikipedia as wp
 
#Get the html source
html = wp.page("List of S&P 500 companies").html().encode("UTF-8")
df = pd.read_html(html)[0]
```

```python
Sector_dic={}
df = df[['Symbol', 'GICS Sector']]
df['Symbol'] = '$' + df.loc[:, 'Symbol']
df = df.set_index('GICS Sector')

for i in df.index:
    Sector_dic[i] = df.loc[i].Symbol.to_list()
    
Sector_dic
```
